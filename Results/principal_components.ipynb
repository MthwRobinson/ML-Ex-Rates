```r
setwd('C:/Users/Eddie/Documents/GitHub/ML-Ex-Rates/Results')

principal_components <- function(country) {
    
    full <- read.csv(paste(country,'_full.csv',sep=''))
    full <- full[,-1]
    rownames(full) <- 1:nrow(full)
    
    prinComps <- princomp(full, cor=TRUE)

    summary(prinComps)
   
}
print("cananda")
principal_components("canada")

print("europe")
principal_components("europe")

print("mexico")
principal_components("mexico")

print("japan")
principal_components("japan")

print("korea")
principal_components("korea")
```
