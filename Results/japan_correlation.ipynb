```r
library(ggplot2)

norm <- function(x) {
    return ((x - min(x)) / (max(x) - min(x)))
}

japan <- read.csv("japan_full.csv")

BOTDomestic <- norm(japan$BOT_d)
FETForeign <- norm(japan$FER_f)


plot(BOTDomestic,type="l",col="red")
plot(FETForeign,type="l",col="green")

```

```r
pdf('japan_plot.pdf')
ggplot() +
    geom_line(data = japan, aes(x = 1:429, y = norm(BOT_d)), colour='red') +
    geom_line(data = japan, aes(x = 1:429, y = norm(FER_f)), colour='green') +
    ggtitle('Balance of Trade and Foreign Exchange Reserves -- Japan') +
    xlab('Month') +
    ylab('Normalized Value')
dev.off()
```
